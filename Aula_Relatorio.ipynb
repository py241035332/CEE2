## Variável aleatória discreta
**Def.:** Seja $X$ uma variável aleatória (v.a.) discreta e $x_1, x_2, …$ os possíveis valores de $X$. Denotamos por $P (X=x_i)$ a probabilidade do evento em que a v.a. $X$ assume o valor $x_i$. Assim a cada possível resultado $x_i$, associamos o número $P (x_i) = P(X=x_i)$ que será denominado **probabilidade** de $x_i$.

Desta forma, $P(x_i)$ deve satisfazer as seguintes condições:

* $P (x_i) ∈ [0, 1], i = 1, 2, 3, …$

* $\sum_{i = 1}^∞ P(x_i) = 1$

**Def.:** A coleção de pares $(x_i, P(x_i))$ para $i = 1, 2, … $ é chamada de **distribuição de probabilidade** da v.a. $X$.

### Distribuição Binomial

A **distribuição Binomial** é uma distribuição de probabilidade discreta que modela o número de sucessos em $n$ experimentos independentes, onde cada experimento tem dois possíveis resultados (sucesso ou fracasso).

A probabilidade de sucesso em um único experimento é denotada por $p$ e a probabilidade de fracasso é $1 − p$.

A fórmula da função de probabilidade para a distribuição binomial é dada por:

$$
P [ X = k ] =\binom{n}{k}p^k(1 − p)^{n − k}, k = 0, 1, 2, …, n.
$$

#### Definições e Propriedades
Notação:
$$
X ∼ Binomial(n,p)
$$

Possíveis valores de $X:0,1,2,…,n$;

#### Valor Esperado

$$
E[X]=np
$$

#### Variância

$$
Var[X]=np(1−p)
$$

#### Gráfico da Distribuição
Abaixo apresentamos o gráfico da função de probabilidade da distribuição binomial com $n =10$ e $p =0.3$.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parâmetros da distribuição
n = 10  # Número de experimentos
p = 0.3  # Probabilidade de sucesso

# Valores possíveis de X
x = np.arange(0, n+1)

# Função de probabilidade
prob = binom.pmf(x, n, p)

# Plotar o gráfico
plt.figure(figsize=(4, 3))
plt.bar(x, prob, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribuição Binomial (n=10, p=0.3)', fontsize=14)
plt.xlabel('Número de sucessos (k)', fontsize=12)
plt.ylabel('P(X=k)', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
